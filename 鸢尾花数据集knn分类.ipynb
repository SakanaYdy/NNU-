```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
if __name__ == '__main__':
    #加载数据集
    iris = load_iris()
    data = iris.get("data")
    target = iris.get("target")
    
    #将数据集划分为0.8的训练数据集 0.2 的测试数据集
    x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=0)
    print(x_test) 
    print("-------------------")
    print(y_test)
    
    #将最近的点位个数设置为5个 
    #就是通过距离该点最近的五个个体 进行类型判断 哪种占比越大 该个体就被划分为那种个图 
    KNN = KNeighborsClassifier(n_neighbors=5)
    KNN.fit(x_train, y_train)
    train_score = KNN.score(x_train, y_train)
    test_score = KNN.score(x_test, y_test)
    print("模型的准确率：", test_score)
    


    
```

```python
#将最近“邻居”个数调到6 个 ，模型预测准确性达到1 （训练集，测试数据集较少） 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
if __name__ == '__main__':
    #加载数据集
    iris = load_iris()
    data = iris.get("data")
    target = iris.get("target")
    
    #将数据集划分为0.8的训练数据集 0.2 的测试数据集
    x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=0)
    print(x_test) 
    print("-------------------")
    print(y_test)
    
    #将最近的点位个数设置为5个 
    #就是通过距离该点最近的五个个体 进行类型判断 哪种占比越大 该个体就被划分为那种个图 
    KNN = KNeighborsClassifier(n_neighbors=6)
    KNN.fit(x_train, y_train)
    train_score = KNN.score(x_train, y_train)
    test_score = KNN.score(x_test, y_test)
    print("模型的准确率：", test_score)
    

    
    #test数据集检验 将分类类型与实际类型进行比对    
    cnt=0
    for x in x_test:
        prediction =KNN.predict([x]) 
        k = iris.get("target_names")[prediction]
        
        label_idx=y_test[cnt]
        if label_idx==0: 
            label="setosa"
        elif label_idx==1:
            label="versicolor"
        else:
            label="virginica"
        
        cnt+=1
        print(cnt,"花的预测种类为：",k,"花的实际种类为:",label)
```
